```python
sample_data = pd.read_csv('Book1.csv') # reading the file 

rate_history = [] # creating an empty list to add later values of col_2 from the file
for i in sample_data.col_2:
    rate_history.append(i)
# finding maximum, minimum, average, and median of the elements of rate_history list
largest_so_far = rate_history[0]
smallest_so_far = rate_history[0]
sum_of_els = 0
count = 0
for i in rate_history:
    sum_of_els = sum_of_els + i
    count = count + 1
    if largest_so_far < i:
        largest_so_far = i
    if smallest_so_far > i:
        smallest_so_far = i
average = sum_of_els / count

print(f'Maximum value is {largest_so_far}')

print(f'Minimum value is {smallest_so_far}')

print(f'Overall average is {average}')

print(f'Median is {statistics.median(rate_history)}')
# rate_history.sort()
# sorted(rate_history)
# for i in range(184):
#     if i == 182:
#         a = rate_history[i]
#     if i == 183:
#         b = rate_history[i]
# print(a/2+b/2)
# median_value = (rate_history[366/2] + rate_history[(366/2)+1])/2.0
# print(f'Median2 is {median_value}')
```
